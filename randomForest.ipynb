```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

X = np.array([[35, 50000, 'Married', 'Govt'], [45, 80000, 'Unmarried', 'Semi-Govt'],
              [28, 35000, 'Divorced', 'Self-Employed'], [42, 60000, 'Married', 'Semi-Govt'],
              [39, 45000, 'Unmarried', 'Govt'], [31, 55000, 'Unmarried', 'Self-Employed'],
              [48, 75000, 'Married', 'Semi-Govt'], [37, 40000, 'Divorced', 'Govt'],
              [40, 65000, 'Unmarried', 'Semi-Govt'], [33, 55000, 'Married', 'Govt']])

y = np.array(['B', 'N', 'B', 'N', 'B', 'B', 'N', 'B', 'N', 'B'])

X[:, 2] = np.where(X[:, 2] == 'Married', 0, np.where(X[:, 2] == 'Unmarried', 1, 2))
X[:, 3] = np.where(X[:, 3] == 'Govt', 0, np.where(X[:, 3] == 'Semi-Govt', 1, 2))
dataset=np.append(X,y.reshape(len(X),1),1)
    
```

```python
dataset
```

```python
def sample(dataset,percent):
  p=list()
  l=list(range(len(dataset)))
  t=np.random.choice(l, size=int(len(dataset)*percent/100), replace=True)
  for i in t:
    p.append(dataset[i])
  p=np.array(p)
  return p
```

```python
models=list()
for i in range(3):
  ds=sample(dataset,70)
  dt=DecisionTreeClassifier(max_depth=3)
  dt.fit(ds[:,0:4],ds[:,4])
  models.append(dt)
```

```python
X_test=np.array([42,60000,0,1])
y_pred=list()
for i in models:
  y_pred.append(list(i.predict(X_test.reshape(1,-1))))
```

```python
y_pred
```

```python
def voting(y):
  y=np.array(y)
  temp=y.flatten()
  b=np.count_nonzero(temp == 'B')
  n=np.count_nonzero(temp == 'N')
  if b>n:
    return 'B'
  else:
    return 'N'
```

```python
voting(y_pred)
```
